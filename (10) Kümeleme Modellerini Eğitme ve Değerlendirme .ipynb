# Kümeleme

*Denetimli* makine öğreniminin aksine, *denetimsiz* öğrenme, etiket tahminlerini eğitmek ve doğrulamak için "temel gerçek" olmadığında kullanılır. Denetimsiz öğrenmenin en yaygın biçimi, kavramsal olarak *sınıflandırmaya* benzeyen *kümelemedir*, ancak eğitim verilerinin tahmin edilecek sınıf etiketi için bilinen değerleri içermemesi dışında. Kümeleme, eğitim durumlarını özellik değerlerinden belirlenebilen benzerliklere göre ayırarak çalışır. Belirli bir varlığın sayısal özellikleri, varlığın n-boyutlu uzaydaki konumunu tanımlayan vektör koordinatları olarak düşünülebilir. Bir kümeleme modelinin yapmaya çalıştığı şey, diğer kümelerden ayrılırken birbirine yakın olan varlıkların gruplarını veya *kümelerini* belirlemektir.

Örneğin, farklı buğday tohumu türlerinin ölçümlerini içeren bir veri kümesine göz atalım.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("seeds.csv")
```

```python
features = df[df.columns[0:6]]
features.sample(10)
```

Gördüğünüz gibi, veri kümesi bir tohumun her örneği (*gözlem*) için altı veri noktası (veya *özellik*) içerir. Yani bunları altı boyutlu uzayda her örneğin konumunu tanımlayan koordinatlar olarak yorumlayabilirsiniz.

Şimdi, elbette altı boyutlu uzayı üç boyutlu bir dünyada veya iki boyutlu bir çizimde görselleştirmek zordur; bu nedenle, özellikler arasındaki ilişkileri analiz etmek ve her gözlemi iki temel bileşen için koordinatlar olarak özetlemek için *Temel Bileşen Analizi* (PCA) adlı bir matematiksel teknikten yararlanacağız - başka bir deyişle, altı boyutlu özelliği tercüme edeceğiz.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
# Sayısal özellikleri aynı ölçekte olacak şekilde normalleştirin
scaled_features = MinMaxScaler().fit_transform(features[df.columns[0:6]])

# İki ana bileşen edinin
pca = PCA(n_components=2).fit(scaled_features)
features_2d = pca.transform(scaled_features)
features_2d[0:10]
```

Artık veri noktalarını iki boyuta çevirdiğimize göre, bunları bir grafikte görselleştirebiliriz:

```python
plt.scatter(features_2d[:,0], features_2d[:,1], color="#9bdc40")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.title("Data")
plt.show()
```

Umarız en az iki, muhtemelen üç, oldukça farklı veri noktası grubu görebilirsiniz; ancak burada kümeleme ile ilgili temel sorunlardan biri yatmaktadır - bilinen sınıf etiketleri olmadan, verilerinizi kaç kümeye ayıracağınızı nasıl bilebilirsiniz?

Bulmaya çalışabileceğimiz bir yol, artan sayıda küme içeren bir dizi kümeleme modeli oluşturmak için bir veri örneği kullanmak ve veri noktalarının her bir küme içinde ne kadar sıkı gruplandığını ölçmektir. Bu sıkılığı ölçmek için sıklıkla kullanılan bir ölçü, *küme içi kareler toplamıdır* (WCSS), daha düşük değerler, veri noktalarının daha yakın olduğu anlamına gelir. Daha sonra her model için WCSS'yi çizebilirsiniz.

```python
from sklearn.cluster import KMeans
```

```python
# 1 ila 10 küme ile 10 model oluşturun
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    # Veri noktalarını sığdırın
    kmeans.fit(features.values)
    # WCSS değerini alın
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11), wcss, color="#e524b6")
plt.title("WCSS by Cluesters")
plt.xlabel("Number of Cluesters")
plt.ylabel("WCSS")
plt.show()
```

Grafik, küme sayısı birden ikiye çıktıkça WCSS'de büyük bir azalma (yani daha büyük *sıkılık*) ve ikiden üç kümeye daha belirgin bir azalma gösteriyor. Bundan sonra, azalma daha az belirgindir ve grafikte yaklaşık üç kümede bir "dirsek" ile sonuçlanır. Bu, makul ölçüde iyi ayrılmış iki ila üç veri noktası kümesi olduğunun iyi bir göstergesidir.

## K-Kmeans Clustering

Test kümelerimizi oluşturmak için kullandığımız algoritma *K-Means*. Bu, bir veri kümesini eşit varyansa sahip *K* kümelere ayıran, yaygın olarak kullanılan bir kümeleme algoritmasıdır. Küme sayısı *K*, kullanıcı tanımlıdır. Temel algoritma aşağıdaki adımlara sahiptir:

1. Bir dizi K ağırlık merkezi rastgele seçilir.
2. Kümeler, veri noktalarının en yakın ağırlık merkezlerine atanmasıyla oluşturulur.
3. Her bir kümenin ortalaması hesaplanır ve ağırlık merkezi ortalamaya taşınır.
4. Adım 2 ve 3, bir durdurma kriteri sağlanana kadar tekrarlanır. Tipik olarak, algoritma, her yeni yineleme, merkezlerin ihmal edilebilir hareketine neden olduğunda ve kümeler statik hale geldiğinde sona erer.
5. Kümeler değişmeyi bıraktığında, algoritma kümelerin konumlarını tanımlayarak *birleşir* - merkezler için rastgele başlangıç ​​noktasının, algoritmayı yeniden çalıştırmanın biraz farklı kümelerle sonuçlanabileceği anlamına geldiğini unutmayın, bu nedenle eğitim genellikle birden çok içerir yinelemeler, merkezler her seferinde yeniden başlatılır ve en iyi WCSS'ye sahip model seçilir.

K değeri 3 olan tohum verilerimizde K-Means kullanmayı deneyelim.

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3, init="k-means++", n_init=100, max_iter=1000)
# init='k-means++', küme merkezlerinin başlangıçta nasıl seçileceğini belirtir.
# n_init=100, modelin farklı başlangıç noktalarından başlayarak en iyisini bulmak için kaç kez çalışacağını belirtir.
# max_iter=1000, maksimum iterasyon sayısını belirtir, yani model, kümeleme işlemini en fazla bu kadar adımda tamamlayacaktır.

km_clusters = model.fit_predict(features.values)
```

İki boyutlu veri noktalarına sahip küme atamalarını görelim.

```python
def plot_clusters(samples, clusters):
    col_dic = {0 : "#80ddfb", 1 : "#9c0d7a", 2 : "#74f530"}
    mrk_dic = {0 : "*", 1 : "o", 2 : "+"}
    colors = [col_dic[x] for x in clusters]
    markers = [mrk_dic[x] for x in clusters]
    for sample in range(len(clusters)):
        plt.scatter(samples[sample][0], samples[sample][1], color=colors[sample], marker=markers[sample], s=100)
    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")
    plt.title("Assignments")
    plt.show()    
```

```python
plot_clusters(features_2d, km_clusters)    
```

Umarım, veriler üç farklı kümeye ayrılmıştır.

Öyleyse kümelemenin pratik kullanımı nedir? Bazı durumlarda, kaç tane küme olduğunu veya bunların neyi gösterdiğini bilmeden farklı kümeler halinde gruplandırmanız gereken verileriniz olabilir. Örneğin, bir pazarlama organizasyonu müşterileri farklı segmentlere ayırmak ve ardından bu segmentlerin nasıl farklı satın alma davranışları sergilediğini araştırmak isteyebilir.

Bazen kümeleme, bir sınıflandırma modeli oluşturmaya yönelik ilk adım olarak kullanılır. Farklı veri noktası gruplarını belirleyerek başlayın ve ardından bu kümelere sınıf etiketleri atayın. Daha sonra bu etiketli verileri bir sınıflandırma modeli eğitmek için kullanabilirsiniz.

Tohum verileri söz konusu olduğunda, farklı tohum türleri zaten bilinmektedir ve 0 (*Kama*), 1 (*Rosa*) veya 2 (*Canadian*) olarak kodlanmıştır, dolayısıyla bu tanımlayıcıları karşılaştırmak için kullanabiliriz. 

```python
seed_species = df[df.columns[7]]
plot_clusters(features_2d, seed_species.values)
```

Küme atamaları ve sınıf etiketleri arasında bazı farklılıklar olabilir, ancak K-Means modeli, aynı türün tohumlarının genellikle aynı kümede olması için gözlemleri kümeleme konusunda makul bir iş çıkarmış olmalıdır.

## Hierarchical Clustering

Hiyerarşik kümeleme yöntemleri, K-means yöntemleriyle karşılaştırıldığında daha az dağıtım varsayımı yapar. Bununla birlikte, K-means yöntemleri genellikle daha ölçeklenebilirdir, bazen çok daha fazla.

Hiyerarşik kümeleme, *bölücü* yöntemle veya *toplayıcı* yöntemle kümeler oluşturur. Bölücü yöntem, tüm veri kümesiyle başlayan ve ardından kademeli bir şekilde bölümleri bulan "yukarıdan aşağıya" bir yaklaşımdır. Aglomeratif kümeleme, "aşağıdan yukarıya" bir yaklaşımdır. Bu laboratuvarda, kabaca aşağıdaki gibi çalışan birleştirici kümeleme ile çalışacaksınız:

1. Veri noktalarının her biri arasındaki bağlantı mesafeleri hesaplanır.
2. Noktalar, en yakın komşularıyla ikili olarak kümelenir.
3. Kümeler arasındaki bağlantı mesafeleri hesaplanır.
4. Kümeler ikili olarak daha büyük kümeler halinde birleştirilir.
5. Adım 3 ve 4, tüm veri noktaları tek bir kümede olana kadar tekrarlanır.

Bağlantı işlevi çeşitli şekillerde hesaplanabilir:
- Koğuş bağlantısı, bağlanan kümeler için varyanstaki artışı ölçer,
- Ortalama bağlantı, iki kümenin üyeleri arasındaki ortalama ikili mesafeyi kullanır,
- Tam veya Maksimal bağlantı, iki kümenin üyeleri arasındaki maksimum mesafeyi kullanır.

Bağlantı fonksiyonlarını hesaplamak için birkaç farklı mesafe ölçümü kullanılır:
- Öklid veya l2 mesafesi en yaygın kullanılanıdır. Bu ölçüm, Ward bağlantı yöntemi için yalnızca bir seçimdir.
- Manhattan veya l1 mesafesi aykırı değerlere dayanıklıdır ve başka ilginç özelliklere sahiptir.
- Kosinüs benzerliği, konum vektörleri arasındaki iç çarpımın vektörlerin büyüklüklerine bölümüdür. Bu ölçünün bir benzerlik ölçüsü olduğuna, diğer iki ölçünün ise farklılık ölçüsü olduğuna dikkat edin. Görüntüler veya metin belgeleri gibi verilerle çalışırken benzerlik oldukça yararlı olabilir.
