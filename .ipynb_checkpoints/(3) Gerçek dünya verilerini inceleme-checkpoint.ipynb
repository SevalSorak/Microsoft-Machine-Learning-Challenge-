# Python ile verileri keşfetme - gerçek dünya verileri

### Gerçek dünya veri dağıtımları

Daha önce, öğrenci verilerimiz için notlara baktık ve bu örnekten tüm not popülasyonunun nasıl görünebileceğini tahmin ettik. Gelin hafızamızı tazeleyelim ve bu verilere tekrar bir göz atalım.

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("basics_grades.csv")
```

```python
df.head()
```

```python
df = df.dropna(axis=0, how="any")
```

```python
# Geçmek için gereken notun '60' olduğunu varsayarak kimin geçtiğini hesaplayın
passes = pd.Series(df["Grade"] >=60)

# Pandas veri çerçevesine kimin geçtiğini kaydet
df = pd.concat([df, passes.rename("Pass")], axis=1)

df.head()
```

```python
# Yeniden kullanabileceğimiz bir işlev oluşturun
def show_distribution(var_data):
    # Bu fonksiyon bir dağılım (grafik) yapacak ve onu gösterecektir.
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mod_val = var_data.mode()[0]
    
    print("Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n".format(min_val,
                                                                                             mean_val,
                                                                                             med_val,
                                                                                             mod_val,
                                                                                             max_val))
    
    # 2 alt grafik için bir şekil oluşturun (2 satır, 1 sütun)
    fig, ax = plt.subplots(2, 1, figsize=(10,8))
    
    # Histogramı çizin
    ax[0].hist(var_data, color = "#0abb21")
    ax[0].set_ylabel("Frequency")
    
    # Ortalama, medyan ve mod için satırlar ekleyin
    ax[0].axvline(x = min_val, color = "#750ab2", linestyle = "dashed", linewidth = 2)
    ax[0].axvline(x = mean_val, color = "#bce32f", linestyle = "dashed", linewidth = 2)
    ax[0].axvline(x = med_val, color = "#e48bdc", linestyle = "dashed", linewidth = 2)
    ax[0].axvline(x = mod_val, color = "#6e4ed3", linestyle = "dashed", linewidth = 2)
    ax[0].axvline(x = max_val, color = "#dd31a1", linestyle = "dashed", linewidth = 2)
    
    # Kutu grafiğini çizin
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel("Value")
    
    # Şekle bir başlık ekleyin
    fig.suptitle("Data Distribution")
    
    fig.show()
```

```python
show_distribution(df["Grade"])
```

Hatırlayacağınız gibi, verilerimizin ortalaması ve modu merkezdeydi ve veriler oradan simetrik olarak dağılmıştı.

Şimdi ders çalışma saatleri verilerinin dağılımına bir göz atalım.

```python
show_distribution(df["StudyHours"])
```

Çalışma süresi verilerinin dağılımı, notlardan önemli ölçüde farklıdır.

Kutu grafiğinin bıyıklarının yalnızca 6.0 civarında başladığına dikkat edin, bu da verilerin ilk çeyreğinin büyük çoğunluğunun bu değerin üzerinde olduğunu gösterir. Minimum, **o** ile işaretlenmiştir ve bu, bunun istatistiksel olarak bir *aykırı* olduğunu, yani dağılımın geri kalanının aralığının önemli ölçüde dışında kalan bir değer olduğunu gösterir.

Aykırı değerler birçok nedenden dolayı ortaya çıkabilir. Belki bir öğrenci "10" saatlik çalışma süresini kaydetmek istedi, ancak "1" girdi ve "0"ı kaçırdı. Ya da iş ders çalışmaya gelince öğrenci anormal derecede tembeldi! Her iki durumda da, tipik bir öğrenciyi temsil etmeyen istatistiksel bir anormallik. Onsuz dağılımın nasıl göründüğünü görelim.

```python
show_distribution(df[df["StudyHours"] > 1]["StudyHours"])
```

Öğrenme amacıyla, **1** değerini burada gerçek bir aykırı değer olarak ele aldık ve hariç tuttuk. Gerçek dünyada, örneklem büyüklüğümüz çok küçükken, aşırı uçlardaki verileri daha fazla gerekçe göstermeden hariç tutmak olağandışı olurdu. Bunun nedeni, örneklem büyüklüğümüz ne kadar küçük olursa, örneklememizin tüm popülasyonu kötü bir şekilde temsil etme olasılığı o kadar yüksektir. (Burada nüfus, sadece 22 öğrencimiz için değil, tüm öğrenciler için not anlamına gelir.) Örneğin, çalışma süresini başka bir 1000 öğrenci için örnek alırsak, aslında fazla çalışmamanın oldukça yaygın olduğunu görebiliriz!

Daha fazla veriye sahip olduğumuzda, örneğimiz daha güvenilir hale gelir. Bu, aykırı değerlerin, verilerin çoğunun içinde yer aldığı yüzdelik dilimlerin altına veya üstüne düşen değerler olarak değerlendirilmesini kolaylaştırır. Örneğin, aşağıdaki kod, 0,01'inci yüzdelik dilimin (verilerin %99'unun üzerinde bulunduğu değer) altındaki gözlemleri hariç tutmak için Pandas **kuantil** işlevini kullanır.

```python
# 0.01. yüzdelik dilimini hesaplayın
q01 = df["StudyHours"].quantile(0.01)

# İncelenecek değişkeni alın
col = df[df["StudyHours"] > q01]["StudyHours"]

# İşlevi çağırın
show_distribution(col)
```

> **İpucu**: Yüksek bir yüzdelik değerde bir eşik tanımlayarak, dağılımın üst ucundaki aykırı değerleri de ortadan kaldırabilirsiniz. Örneğin, verilerin %99'unun altında bulunduğu 0,99 yüzdelik dilimini bulmak için **nicelik** işlevini kullanabilirsiniz.

Aykırı değerler kaldırıldığında, kutu grafiği dört çeyrek içindeki tüm verileri gösterir. Dağılımın, derece verileri için olduğu gibi simetrik olmadığına dikkat edin. Yaklaşık 16 saat gibi çok yüksek çalışma sürelerine sahip bazı öğrenciler var, ancak verilerin büyük kısmı 7 ile 13 saat arasında. Son derece yüksek birkaç değer, ortalamayı ölçeğin üst ucuna doğru çeker.

Bu dağılımın yoğunluğuna bakalım.

```python
def show_density(var_data):
    fig = plt.figure(figsize=(10,8))
    
    # Yoğunluk grafiği
    var_data.plot.density()
    
    # Başlık ve etiket ekleyin
    plt.title("Data Density")
    
    # Ortalamayı, medyanı ve modu gösterin
    plt.axvline(x = var_data.mean(), color = "#97bfa7", linestyle = "dashed", linewidth = 2)
    plt.axvline(x = var_data.median(), color = "#459b34", linestyle = "dashed", linewidth = 2)
    plt.axvline(x = var_data.mode()[0], color = "#b3be67", linestyle = "dashed", linewidth = 2)
    
    plt.show()
```

```python
show_density(col)
```

Bu tür dağılıma *sağa çarpık* denir. Verilerin kütlesi dağılımın sol tarafındadır ve ortalamayı sağa çeken aşırı yüksek uçtaki değerler nedeniyle sağa doğru uzun bir kuyruk oluşturur.

#### Varyans ölçüleri

Şimdi sınıf ortası ve çalışma saatleri veri dağılımlarının nerede olduğu hakkında iyi bir fikrimiz var. Ancak, dağılımların incelememiz gereken başka bir yönü daha var: Verilerde ne kadar değişkenlik var?

Verilerdeki değişkenliği ölçen tipik istatistikler şunları içerir:

- **Aralık**: Maksimum ve minimum arasındaki fark. Bunun için yerleşik bir işlev yoktur, ancak **min** ve **max** işlevlerini kullanarak hesaplaması kolaydır.
- **Varyans**: Ortalamadan karesi alınmış farkın ortalaması. Bunu bulmak için yerleşik **var** işlevini kullanabilirsiniz.
- **Standart Sapma**: Varyansın karekökü. Bunu bulmak için yerleşik **std** işlevini kullanabilirsiniz.

```python
for col_name in ["Grade", "StudyHours"]:
    col = df[col_name]
    rng = col.max() - col.min()
    var = col.var()
    std = col.std()
    print("\n{}:\n - Range: {:.2f}\n - Variance: {:.2f}\n - Std: {:.2f}".format(col_name, rng, var, std))
```

Bu istatistiklerden standart sapma genellikle en kullanışlı olanıdır. Verilerin kendisiyle aynı ölçekte bir varyans ölçüsü sağlar (bu nedenle Not dağılımı için not puanları ve StudyHours dağılımı için saatler). Standart sapma ne kadar yüksek olursa, dağılımdaki değerleri dağılım ortalamasıyla karşılaştırırken o kadar fazla varyans olur, başka bir deyişle, veriler daha fazla dağılır.

*Normal* bir dağılımla çalışırken standart sapma, daha da fazla bilgi sağlamak için normal dağılımın belirli özellikleriyle birlikte çalışır. Standart sapmalar ile normal dağılımdaki veriler arasındaki ilişkiyi görelim.

```python
import scipy.stats as stats
```

```python
# Not sütununu alın
col = df["Grade"]

# Yoğunluğu elde edin
density = stats.gaussian_kde(col)

# Yoğunluğu çizin
col.plot.density()

# Ortalamayı ve standart sapmayı alın
s = col.std()
m = col.mean()

# 1 stdev'e açıklama ekleyin
x1 = [m - s, m + s] # 1 standart sapma aralığı
y1 = density(x1) # x1 aralığındaki yoğunluk değerleri
plt.plot(x1, y1, color = "#7fd2ce")
plt.annotate("1 std (68.26%)", (x1[1], y1[1]))
# x1 ve y1 değerlerini kullanarak 1 std aralığının %68.26'sını ifade eden metni, eğrinin uygun noktasına ekliyoruz.

# 2 stdev'e açıklama ekleyin
x2 = [m - (s * 2), m + (s * 2)] 
y2 = density(x2) 
plt.plot(x2, y2, color = "#cec356")
plt.annotate("2 std (95.45%)", (x2[1], y2[1]))

# 3 stdev'e açıklama ekleyin
x3 = [m - (s * 3), m + (s * 3)] 
y3 = density(x3)
plt.plot(x3, y3, color = "#71d0f8")
plt.annotate("3 std (99.73%)", (x3[1], y3[1]))

# Ortalamanın yerini gösterin
plt.axvline(col.mean(), color = "#227dd6", linestyle = "dashed", linewidth = 1)

plt.axis("off")
plt.show()
```

Yatay çizgiler, ortalamanın (artı veya eksi) 1, 2 ve 3 standart sapma içindeki verilerin yüzdesini gösterir.

Herhangi bir normal dağılımda:
- Değerlerin yaklaşık %68,26'sı ortalamadan bir standart sapma dahilindedir.
- Değerlerin yaklaşık %95,45'i ortalamadan iki standart sapma içindedir.
- Değerlerin yaklaşık %99,73'ü ortalamadan üç standart sapma dahilindedir.

Yani not ortalamasının 49.18, standart sapmasının 21.74 olduğunu ve not dağılımının yaklaşık olarak normal olduğunu bildiğimiz için öğrencilerin %68.26'sının 27.44 ile 70.92 arasında bir not alması gerektiğini hesaplayabiliriz.

Öğrenci veri değişkenlerinin dağılımını anlamak için kullandığımız tanımlayıcı istatistikler, istatistiksel analizin temelini oluşturur. Verilerinizi keşfetmenin çok önemli bir parçası olduklarından, tüm sayısal sütunlar için ana tanımlayıcı istatistikleri döndüren DataFrame nesnesinin yerleşik bir "açıklama" yöntemi vardır.

## Verileri karşılaştırma

Artık veri kümenizdeki verilerin istatistiksel dağılımı hakkında bir şeyler bildiğinize göre, değişkenler arasındaki belirgin ilişkileri belirlemek için verilerinizi incelemeye hazırsınız.

Her şeyden önce, tipik bir öğrenci sınıfını temsil eden bir örneğe sahip olmak için aykırı değerler içeren tüm satırlardan kurtulalım. StudyHours sütununun son derece düşük değerlere sahip bazı aykırı değerler içerdiğini belirledik, bu nedenle bu satırları kaldıracağız.

```python
df_sample = df[df["StudyHours"] > 1]
df_sample.head()
```

Bu karşılaştırmayı yapmak için, her olası Geçiş değeri (doğru ve yanlış) için StudyHours dağılımını gösteren kutu çizimleri oluşturalım.

### Sayısal değişkenleri karşılaştırma

Şimdi iki sayısal değişkeni karşılaştıralım. Hem notu hem de çalışma saatlerini gösteren bir çubuk grafik oluşturarak başlayacağız.

```python
df_sample.plot(x = "Name", y = ["Grade", "StudyHours"], kind = "bar", figsize=(8,5), color = ["#03a072", "#df9b6b"])
```

Grafik, her öğrenci için hem sınıf hem de çalışma saatleri için çubuklar gösterir, ancak değerler farklı ölçeklerde olduğu için karşılaştırmak kolay değildir. Not, not puanlarıyla ölçülür (ve 3 ile 97 arasında değişir) ve çalışma süresi saatlerle ölçülür (ve 1 ile 16 arasında değişir).

Farklı ölçeklerdeki sayısal verilerle uğraşırken yaygın olarak kullanılan bir teknik, değerlerin orantılı dağılımlarını koruyacak ancak aynı ölçekte ölçülecek şekilde verileri *normalleştirmek*tir. Bunu başarmak için, değerleri 0'dan 1'e orantılı olarak dağıtan *MinMax* ölçeklendirme adlı bir teknik kullanacağız. Bu dönüşümü uygulamak için kod yazabilirsiniz, ancak **Scikit-Learn** kitaplığı bunu sizin için yapacak bir ölçekleyici sağlar.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

df_normalized = df_sample[["Name", "Grade", "StudyHours"]].copy()

df_normalized[["Grade", "StudyHours"]] = scaler.fit_transform(df_normalized[["Grade", "StudyHours"]])

df_normalized.plot(x = "Name", y = ["Grade", "StudyHours"], kind = "bar", figsize=(8,5), color = ["#03a072", "#df9b6b"])
```

Veriler normalleştirildiğinde, not ve çalışma süresi arasında bariz bir ilişki görmek daha kolaydır. Tam bir eşleşme değil, ama kesinlikle daha yüksek notlara sahip öğrencilerin daha fazla çalışma eğiliminde olduğu görülüyor.

Yani çalışma süresi ve not arasında bir korelasyon var gibi görünüyor. Aslında, bu sütunlar arasındaki ilişkiyi ölçmek için kullanabileceğimiz istatistiksel bir *korelasyon* ölçümü vardır.

```python
df_normalized.Grade.corr(df_normalized.StudyHours)
```

Korelasyon istatistiği, bir ilişkinin gücünü gösteren -1 ile 1 arasında bir değerdir. 0'ın üzerindeki değerler *pozitif* bir korelasyonu gösterir (bir değişkenin yüksek değerleri diğerinin yüksek değerleri ile çakışma eğilimindedir), 0'ın altındaki değerler *negatif* bir korelasyonu gösterir (bir değişkenin yüksek değerleri diğerinin düşük değerleri ile çakışma eğilimindedir). Bu durumda, korelasyon değeri 1'e yakındır ve çalışma süresi ile derece arasında güçlü bir pozitif korelasyon gösterir.

> **Not**: Veriler öğrenci verilerinin durumudur, istatistikler notları yüksek olan öğrencilerin aynı zamanda yüksek miktarda çalışma süresine sahip olduklarını göstermektedir, ancak bu onların *çok çalıştıkları için* yüksek notlar aldıklarını kanıtlamakla aynı şey değildir. İstatistik, öğrencilerin notlarının yüksek olacağı *çünkü* çok çalıştıkları şeklindeki saçma sapan sonucu desteklemek için eşit derecede kanıt olarak kullanılabilir.

İki sayısal sütun arasındaki bariz korelasyonu görselleştirmenin başka bir yolu da bir *dağılım* grafiği kullanmaktır.

```python
df_sample.plot.scatter(title = "Study Time vs Grade", x = "StudyHours", y = "Grade", color = "#4b310c")
```

Yine, en çok saat çalışan öğrencilerin aynı zamanda en yüksek notları alan öğrenciler olduğu fark edilebilir bir model var gibi görünüyor.

Verilerdeki genel eğilimi gösteren grafiğe bir *gerileme* çizgisi (veya *en uygun çizgi*) ekleyerek bunu daha net görebiliriz. Bunu yapmak için *en küçük kareler regresyonu* adı verilen istatistiksel bir teknik kullanacağız.

Okulda lineer denklemleri nasıl çözeceğinizi öğrendiğinizi hatırlayın ve lineer bir denklemin *eğim-kesim noktası* formunun şöyle göründüğünü hatırlayın:

$ y = mx + b $

Bu denklemde, *y* ve *x* koordinat değişkenleridir, *m* doğrunun eğimidir ve *b* y-kesme noktasıdır (doğrunun Y ekseninden geçtiği yer).

Öğrenci verilerimiz için dağılım grafiğimizde, *x* (*ÇalışmaSaatleri*) ve *y* (*Sınıf*) için değerlerimize zaten sahibiz, dolayısıyla bu noktalara en yakın düz çizginin kesişimini ve eğimini hesaplamamız yeterli. Ardından, *x* (*StudyHours*) değerlerimizin her biri için o satırda yeni bir *y* değeri hesaplayan bir doğrusal denklem oluşturabiliriz. Karışıklığı önlemek için, bu yeni *y* değerini *f(x)* olarak adlandıracağız (çünkü bu, *x*'e dayalı doğrusal bir denklemin ***f*** işleminin çıktısıdır). Orijinal *y* (*Not*) değeri ile *f(x)* değeri arasındaki fark, regresyon doğrumuz ile öğrencinin ulaştığı gerçek *Not* arasındaki *hatadır*. Amacımız, en düşük toplam hataya sahip bir çizgi için eğimi ve kesişimi hesaplamaktır.

Spesifik olarak, her nokta için hatayı alıp karesini alarak ve karesi alınan tüm hataları toplayarak genel hatayı tanımlarız. En uygun çizgi, bize kareleri alınmış hataların toplamı için en düşük değeri veren çizgidir; bu nedenle *en küçük kareler regresyonu* adı verilir.

Neyse ki, regresyon hesaplamasını kendiniz kodlamanıza gerek yok. **SciPy** paketi, zorlu işleri sizin yerinize yapacak bir **linregress** yöntemi sağlayan bir **stats** sınıfı içerir. Bu, (diğer şeylerin yanı sıra) eğim denklemi için ihtiyaç duyduğunuz katsayıları döndürür: karşılaştırmak istediğiniz belirli bir değişken örnek çiftine dayalı olarak eğim (*m*) ve kesişme noktası (*b*).

```python
df_regression = df_sample[["Grade", "StudyHours"]].copy()

# Regresyon eğimini alın ve kesiştirin
m, b, r, p, se = stats.linregress(df_regression["StudyHours"], df_regression["Grade"])
print("slope: {:.4f}\ny-intercept: {:.4f}".format(m,b))
print("so...\n f(x) = {:.4f}x + {:.4f}".format(m,b))

# Her x (StudyHours) değeri için f(x)'i hesaplamak üzere (mx + b) işlevini kullanın
df_regression["fx"] = (m * df_regression["StudyHours"]) + b

# f(x) ile gerçek y (Derece) değeri arasındaki hatayı hesaplayın
df_regression["error"] = df_regression["fx"] - df_regression["Grade"]

# Grade ve StudyHours'un dağılım grafiğini oluşturun
df_regression.plot.scatter(x = "StudyHours", y = "Grade", color = "#aa6b5e")

# Regresyon çizgisini çizin
plt.plot(df_regression["StudyHours"], df_regression["fx"], color = "#a631a3")

plt.show()
```

Bu kez kodun iki farklı şeyi çizdiğine dikkat edin: Örnek çalışma saatlerinin ve notların dağılım grafiği önceki gibi çizilir ve ardından en küçük kareler regresyon katsayılarına dayalı en uygun çizgi çizilir.

Regresyon çizgisi için hesaplanan eğim ve kesişme katsayıları grafiğin üzerinde gösterilmiştir.

Satır, her bir **StudyHours** değeri için hesaplanan ***f*(x)** değerlerini temel alır. 

- Her öğrenci için **Çalışma Saatleri**.
- Her öğrencinin ulaştığı **Not**.
- Regresyon çizgisi katsayıları kullanılarak hesaplanan ***f(x)*** değeri.
- Hesaplanan ***f(x)*** değeri ile gerçek **Grade** değeri arasındaki *hata*.

Hatalardan bazıları, özellikle aşırı uçlarda ve oldukça büyük (17,5 puanın üzerine kadar). Ancak, genel olarak, çizgi gerçek notlara oldukça yakındır.

### Tahmin için regresyon katsayılarını kullanma

Artık çalışma süresi ve sınıf ilişkisi için regresyon katsayılarına sahip olduğunuza göre, bunları belirli bir çalışma miktarı için beklenen notu tahmin eden bir işlevde kullanabilirsiniz.

```python
# Regresyon katsayılarımıza dayalı olarak bir fonksiyon tanımlayın
def f(x):
    m = 6.3134
    b = -17.9164
    return m*x + b

study_time = 14

# Çalışma süresi için f(x) alın
prediction = f(study_time)

# Not 0'dan az ve 100'den fazla olamaz
expected_grade = max(0, min(100, prediction))

#Tahmini notu yazdırın
print("Haftada {} saat çalışmak {:.0f} notuyla sonuçlanabilir.".format(study_time, expected_grade))
```

İstatistikleri örnek verilere uygulayarak, çalışma süresi ile not arasında bir ilişki belirlediniz ve bu ilişkiyi, belirli bir çalışma süresi için bir notu tahmin etmek için kullanılabilecek genel bir fonksiyonda özetlediniz.

Bu teknik aslında makine öğreniminin temel dayanağıdır. Bir veya daha fazla *özelliği* (bu durumda çalışılan saat sayısı) ve bilinen bir *etiket* değerini (bu durumda elde edilen notu) içeren bir örnek veri seti alabilir ve verilen herhangi bir özellik grubu için tahmin edilen etiket değerlerini hesaplayan bir fonksiyon türetmek için örnek verileri kullanabilirsiniz.
